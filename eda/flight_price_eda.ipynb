```python
#importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
df=pd.read_excel('flight_price.xlsx')
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
#all the converted columns are in object so changing them into int
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
#drop the data of journey column
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
#Extract only the time part first
df['Arrival_time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])

# Now split hour and minute from the cleaned time
df['Arrival_Hour'] = df['Arrival_time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_time'].str.split(':').str[1]
```

```python
df.head()
```

```python
#converting th data type of Arrival_Hour and Arrival_Minute to int
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
df.head()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
df.head()
```

```python
#Now splitting the detail of the departure time
df['Departure_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_Mintue']=df['Dep_Time'].str.split(':').str[1]
df.head()
```

```python
#converting th departure hour and minute to int 
df['Departure_Hour']=df['Departure_Hour'].astype(int)
df['Departure_Mintue']=df['Departure_Mintue'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
#checking the values of total_stops
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Extract hour and minute as numbers from 'Duration'
df['Duration_Hour'] = df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)
df['Duration_Minute'] = df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int)
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
onehot_df=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
# Concatenate one-hot encoded columns to the main DataFrame
df = pd.concat([df.reset_index(drop=True), onehot_df.reset_index(drop=True)], axis=1)
```
