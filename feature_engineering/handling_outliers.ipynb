```python
#Handling our outliers is one of the major setps in feature engineering.
#This script will help you to identify and handle outliers in your dataset.
```

```python
import numpy as np
```

```python
lst_marks=np.random.randint(0,1001,20)
print("Marks:",lst_marks)
mini,q1,median,q3,max=np.quantile(lst_marks,[0,0.25,0.50,0.75,1])
```

```python
IQR=q3-q1
print("IQR:",IQR)
```

```python
lower_fence=q1-1.5*IQR
upper_fence=q3+1.5*IQR
print("Lower Fence:",lower_fence)
print("Upper Fence:",upper_fence)
```

```python
import seaborn as sns
sns.boxplot(lst_marks)
```
