```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zomato sales analysis ==== ")
print("\n Sales data shape: ",sales_data.shape)
print("\n Sample data for first 3 resturants: \n",sales_data[0:3]) #getting first thre columns
```

# Total sales per year

```python
yearly=np.sum(sales_data[:,1:],axis=0)
print(yearly)

#minimum sales
minium=np.min(sales_data[:,1:],axis=1)
print(minium)

#maximun sales per year
maximum_sales=np.max(sales_data[:,1:],axis=0)
print(maximum_sales)

#average sales per restureant 
average_sales=np.average(sales_data[:,1:],axis=1)
print(average_sales)

#cumulative sales
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)
```

```python
plt.figure(figsize=(6,4))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all resturant: ")
plt.xlabel("Years")
plt.ylabel("Sales")
```

```python
#calculate the monthly average

monthly_sales=sales_data[:,1:] / 12 
print("The average monthly sales: \n",monthly_sales)
```
