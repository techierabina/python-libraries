```python
#grouping
# Step 1: Find numeric columns in the desired range
numeric_cols = df[df.columns[5:13]].select_dtypes(include='number').columns
# Step 2: Group by Continent, take mean of those columns, sort
df2 = df.groupby('Continent')[numeric_cols].mean().sort_values(by="2022 Population", ascending=False)
df2
```

```python
df3=df2.transpose()
df3
```
