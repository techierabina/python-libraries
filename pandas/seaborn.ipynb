```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#basic plotting with seaborn
tips = sns.load_dataset("tips")
tips
```

```python
##create a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter plot of Total Bill vs Tip")
plt.show()
```

```python
## line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of Size vs Total Bill")
plt.show()
```

```python
## Categorical plot (bar plot)
sns.barplot(x='day',y='total_bill',data=tips )
plt.title("Bar plot of Total Bill by Day")
plt.show()
```

```python
## box plot
sns.boxplot(x='day',y='total_bill', data=tips)
plt.title("Box plot of Day by Total Bill")
plt.show()
```

```python
## Voilin plot 
sns.violinplot(x='day',y='total_bill', data=tips)
plt.title("Violin plot of Day by Total Bill")
plt.show()  
```

```python
## Histogram
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
## KDE plot
sns.kdeplot(tips['total_bill'], shade=True)
plt.title("KDE plot of Total Bill")
plt.show()
```

```python
#Pair plot
sns.pairplot(tips, hue='day')
```

```python
#Heatmap
corr=tips[['total_bill','tip','size']].corr() 
corr
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap of Total Bill, Tip, and Size")
plt.show()
```

```python
## customization
import pandas as pd 
df=pd.read_csv('customer_data.csv')
df
```

```python
# Calculate average spending score by country
avg_spending = df.groupby('Country', as_index=False)['SpendingScore'].mean().sort_values(by='SpendingScore', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='SpendingScore', data=avg_spending, palette='viridis')
plt.title('Top 10 Countries by Average Spending Score')

plt.show()
```
